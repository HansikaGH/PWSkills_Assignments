Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Answer:

Five plots in Seaborn along with their usage is as following :

1. Box Plot : A box plot is a graphical representation of data that shows the distribution of a dataset by indicating the median, quartiles, and outliers. 
- use: It is used to identify the spread and skewness of the data.

2. Heatmap : A heatmap is a graphical representation of data that uses color-coding to represent the values of a two-dimensional dataset. 
- use: It is used to identify patterns in the data, such as correlations or clusters.

3. Pair Plot : A pair plot is a graphical representation of data that shows pairwise relationships between different variables in a dataset. 
- use: It is used to identify correlations between variables and to explore the relationships between them.

4. Violin Plot : A violin plot is a graphical representation of data that combines elements of a box plot and a kernel density plot. 
- use: It is used to visualize the distribution of a dataset and show its density.
5. Joint Plot : A joint plot is a graphical representation of data that shows the relationship between two variables using a scatter plot and a histogram. 
- use: It is used to explore the relationship between two variables and to identify correlations between them.

These are just a few examples of the types of plots that can be created using Seaborn. Seaborn offers many other visualizations that can be used to explore and analyze different types of data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Sample data
tips = sns.load_dataset("tips")
```

```python
##1. Box Plot

# Box plot of tips by day
sns.boxplot(x="day", y="tip", data=tips)
plt.title("Tip Distribution by Day")
plt.show()
```

```python
##2. Heatmap

# Sample data: correlation matrix
iris = sns.load_dataset("iris")
corr = iris.corr(numeric_only=True)

# Heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
##3. Pairplot
# Pairplot of iris dataset
sns.pairplot(iris, hue="species")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()

```

```python
##4. Violin Plot
# Violin plot of petal length by species
sns.violinplot(x="species", y="petal_length", data=iris)
plt.title("Violin Plot of Petal Length by Species")
plt.show()
```

```python
##5. Joint Plot

# Joint plot of sepal length vs petal length
sns.jointplot(x="sepal_length", y="petal_length", data=iris, kind="scatter")
plt.suptitle("Joint Plot of Sepal vs Petal Length", y=1.02)
plt.show()
```

Question 6 : Use the "flights" dataset from seaborn to plot a heatmap

Answer :

Importing flights dataset

```python
import seaborn as sns
flights = sns.load_dataset('flights')
flights.head()
```

Pivot the dataset to create a matrix with month on the x-axis, year on the y-axis, and passengers as values

```python
flights_matrix = flights.pivot(index="year", columns="month", values="passengers")
flights_matrix
```

Creating heatmap using Seaborn

```python
import matplotlib.pyplot as plt
sns.heatmap(flights_matrix, 
            cmap="YlGnBu", 
            annot=True, 
            fmt="d", 
            linewidths=.5,
            linecolor="black")
plt.title("Passengers per Month")
plt.show()
```
